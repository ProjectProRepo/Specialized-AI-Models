```python


# Imports
from transformers import pipeline
import torch
from IPython.display import FileLink

# Load text file
def load_ticket(file_path):
    with open(file_path, "r") as f:
        return f.read()

# Summarize complaint or ticket
def summarize_ticket(text, model_name="Falconsai/text_summarization"):
    print("Loading summarization model...")
    summarizer = pipeline("summarization", model=model_name)
    summary = summarizer(text, max_length=150, min_length=40, do_sample=False)[0]['summary_text']
    return summary

# Main handler
def run_ticket_summarizer(ticket_file):
    ticket = load_ticket(ticket_file)
    print("\n--- Full Ticket ---\n", ticket)

    summary = summarize_ticket(ticket)
    print("\n--- Summary ---\n", summary)

    with open("ticket_summary.md", "w") as f:
        f.write("# Ticket Summary\n\n")
        f.write(summary)
        f.write("\n\n---\n\n")
        f.write("## Full Log\n\n")
        f.write(ticket)

# Run
if __name__ == "__main__":
    ticket_path = "/content/complaint_log.txt"  # change filename if needed
    run_ticket_summarizer(ticket_path)

# Download the summary
FileLink("/content/ticket_summary.md")
```
